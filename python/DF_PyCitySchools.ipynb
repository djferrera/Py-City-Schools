```python
# import modules / dependencies
import pandas as pd
import numpy as np
```

```python
# load data / csv files - school
school_data = "schools_complete.csv"

# make a reference to the budget_data.csv file path
csv_path_1 = "schools_complete.csv"

# import the csv files as a dataframes
schools_df = pd.read_csv(csv_path_1, encoding="utf-8")
```

```python
# load data / csv files - student
student_data = "students_complete.csv"

# make a reference to the budget_data.csv file path
csv_path_2 = "students_complete.csv"

# import the csv files as a dataframes
students_df = pd.read_csv(csv_path_2, encoding="utf-8")
```

```python
# show students column headers
students_df.columns
```

```python
# need to merge tables together to get one data set
# outer join keeps all data
education_df = pd.merge(schools_df, students_df, on="school_name", how="outer")
```

```python
# show education column headers
education_df.columns
```

```python
# Assignment 1: District Summary
# Create a high level snapshot (in table form) of the districtâ€™s key metrics, including:
# Total Schools, Total Students, Total Budget, Average Math Score, Average Reading Score,
# % Passing Math, % Passing Reading, Overall Passing Rate (Average of the above two)
# format style at end!
```

```python
# list of schools
total_schools = education_df["school_name"].unique()

# total number of schools
total_schools_count = total_schools.shape[0]
```

```python
# list of students
total_students = education_df["student_name"].count()
```

```python
# unique school budgets *does not account for schools with same budget!**
# need to use a better method
school_budgets = education_df["budget"].unique()

# total budget
total_budget = school_budgets.sum()
```

```python
# average math score
average_math = education_df["math_score"].mean()
```

```python
# average reading score
average_read = education_df ["reading_score"].mean()
```

```python
# students who passed math - passing is 70
pass_math = education_df.loc[education_df['math_score'] > 69].count()

#percentage of students who passed math
perc_pass_math = (pass_math / total_students)[0]*100
```

```python
# students who passed reading - passing is 70
pass_read = education_df.loc[education_df['reading_score'] > 69].count()

#percentage of students who passed math
perc_pass_read = (pass_read / total_students)[0]*100
```

```python
# passing rate (overall average score)
# i.e. (avg. math score + avg. reading score)/2
passing_rate = (average_math + average_read)/2
```

```python
# create district summary table
district_summary_df = pd.DataFrame(
        {"Total Schools": [total_schools_count], "Total Students": [total_students],
        "Total Budget": [total_budget], "Average Math Score": [average_math],
        "Average Reading Score": [average_read], "% Passing Math": [perc_pass_math],
        "% Passing Reading": [perc_pass_read], "% Overall Passing Rate": [passing_rate]})
district_summary_df
```

```python
# format district summary table
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.2f}%".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.2f}%".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.2f}%".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.2f}%".format)
district_summary_df["% Overall Passing Rate"] = district_summary_df["% Overall Passing Rate"].map("{:.2f}%".format)
district_summary_df
```

```python
# Assignment 2: School Summary
# Overview table that summarizes key metrics about each school, including:
# School Name, School Type, Total Students, Total School Budget, Per Student Budget, Average Math Score,
# Average Reading Score, % Passing Math, % Passing Reading, Overall Passing Rate (Average of the above two)
```

```python
# GroupBy "school_name"
per_school = education_df.groupby(['school_name'])
```

```python
# budget per school
school_budget = per_school["budget"].mean()
```

```python
# per student budget
student_budget = school_budget / school_students
```

```python
# average math score per school
avg_math = per_school["math_score"].mean()
```

```python
# average reading score per school
avg_read = per_school["reading_score"].mean()
```

```python
# % passing math (70%)               
math_pass = education_df.loc[education_df['math_score'] > 69].groupby(["school_name"]).size()

# percentage of students per school who passed math
per_math_pass = (math_pass / school_students)*100
```

```python
# % passing reading (70%)               
read_pass = education_df.loc[education_df['reading_score'] > 69].groupby(["school_name"]).size()

# percentage of students per school who passed math
per_read_pass = (read_pass / school_students)*100
```

```python
# overall passing per school
school_pass_rate = (per_math_pass + per_read_pass)/2
```

```python
# new data frame based on school_summary_df
spending_df = school_summary_raw.loc[:,['Average Math Score','Average Reading Score','% Passing Math', '% Passing Reading','% Overall Passing Rate','Per Student Budget']]
```
